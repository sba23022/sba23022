```python
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd
import re
import nltk

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Assuming file_content is a string containing all the reviews
# Split the content into individual reviews
# Assuming each review is on a new line
Transport_reviews = file_content.split('\n')

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Get English stopwords
stop_words = set(stopwords.words('english'))

# Function to clean and remove stopwords from a review
def clean_review(review):
    # Remove non-alphabetic characters and lowercase the words
    review = re.sub("[^a-zA-Z]", " ", review).lower()

    # Tokenize the review
    word_tokens = word_tokenize(review)

    # Filter out stopwords
    filtered_review = [word for word in word_tokens if word not in stop_words]

    # Rejoin words to form the cleaned review
    return ' '.join(filtered_review)

# Analyze sentiment for each cleaned review
sentiments = []
for review in Transport_reviews:
    if review:  # Check if the review is not empty
        cleaned_review = clean_review(review)
        score = sia.polarity_scores(cleaned_review)
        sentiments.append(score)

# Create a DataFrame
sentiment_analysis = pd.DataFrame(sentiments)

# Display the sentiment scores
print(sentiment_analysis )
```

Hypothesis Probability Question: "Is there a significant difference in the average weekly volume of cars between the first half and the second half of the year 2019 in Dublin, Ireland?"

Rationale: This hypothesis aims to determine if traffic volumes significantly differ between the two halves of the year, suggesting seasonal variation in traffic patterns.

The T-test result, with a T-statistic of -1.7032 and a P-value of 0.0891, is used to assess this hypothesis. The negative T-statistic indicates that the average traffic volume in the first half is lower than in the second half. However, the P-value is above the conventional threshold of 0.05, which suggests that this difference is not statistically significant.

Decision on Hypothesis: We fail to reject the null hypothesis because the P-value is greater than 0.05. This implies that there isn't enough statistical evidence to conclude a significant difference in traffic volumes between the first and second halves of the year. The observed difference might be due to random variations rather than a consistent seasonal effect.

```python
from scipy.stats import ttest_ind, wilcoxon, mannwhitneyu

# Pairwise comparisons for T-Test, Wilcoxon, and Mann-Whitney Tests
pairs = [("Dublin", "Leeds"), ("Dublin", "Edinburgh"), ("Leeds", "Edinburgh")]

# Initializing dictionaries to store test results
ttest_results = {}
wilcoxon_results = {}
mannwhitney_results = {}

# Conducting the tests
for city1, city2 in pairs:
    # Extracting data for each pair
    data1 = anova_data_sampled[anova_data_sampled['City'] == city1]['Average_weekly_volume_of_cars']
    data2 = anova_data_sampled[anova_data_sampled['City'] == city2]['Average_weekly_volume_of_cars']

    # T-Test
    ttest_results[f"{city1} vs {city2}"] = ttest_ind(data1, data2, equal_var=False)

    # Wilcoxon Test - requires equal lengths of data, hence we take a sample of equal size from each
    min_size = min(len(data1), len(data2))
    wilcoxon_results[f"{city1} vs {city2}"] = wilcoxon(data1.sample(min_size), data2.sample(min_size))

    # Mann-Whitney Test
    mannwhitney_results[f"{city1} vs {city2}"] = mannwhitneyu(data1, data2)

# Displaying the results
ttest_results, wilcoxon_results, mannwhitney_results

```

Mann-Whitney Test: Another non-parametric test, used to compare differences between two independent groups. We'll also apply this test pairwise.

```python
import pandas as pd
import requests

# URL of the JSON data
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/THA21/JSON-stat/1.0/en"

# Fetch the data from the URL
response = requests.get(url)
json_dataset = response.json()

# Extract the dataset
dataset = json_dataset['dataset']

# Extract dimensions and their labels
dimensions = dataset['dimension']
dimension_ids = dimensions['id']
labels = {}

# Create label mappings for each dimension
for dim_id in dimension_ids:
    label_data = dimensions[dim_id]['category']
    labels[dim_id] = label_data['label']

# Extract values
values = dataset['value']

# Initialize DataFrame rows
rows = []

# Calculate the number of items for each dimension
n_items = [len(dimensions[dim_id]['category']['index']) for dim_id in dimension_ids]
n_total = len(values)

# Process each value
for idx, value in enumerate(values):
    if value is None:
        continue  # Skip missing values

    # Decompose the index into its components
    indices = []
    product = n_total
    for n in n_items:
        product //= n
        index_key = list(dimensions[dimension_ids[n_items.index(n)]]['category']['index'].keys())[idx // product % n]
        indices.append(index_key)

    # Map indices to labels
    row_labels = [labels[dim_id][index] for dim_id, index in zip(dimension_ids, indices)]

    # Append the row
    rows.append(row_labels + [value])

# Define column names based on dimension IDs and value
column_names = dimension_ids + ['VALUE']

# Create the DataFrame
df = pd.DataFrame(rows, columns=column_names)

# Rename columns to match the Ireland dataset structure
rename_map = {
    'STATISTIC': 'Statistic Label',
    'TLIST(W1)': 'Year',
    'C03910V04662': 'Transport Traffic Site',
    'C01198V01436': 'Weeks of the year'
}
df.rename(columns=rename_map, inplace=True)

# Add 'UNIT' column assuming all values are numbers
df['UNIT'] = 'Number'

# Reorder columns to match the Ireland dataset
df = df[['Statistic Label', 'Year', 'Transport Traffic Site', 'Weeks of the year', 'UNIT', 'VALUE']]

# Display the DataFrame
print(df.head())
```

```python
import pandas as pd
from datetime import datetime, timedelta
import warnings
# Ignoring all warnings
warnings.filterwarnings('ignore')

# Create a function to generate a synthetic school calendar dataset
def create_school_calendar(start_year, end_year):
    # Define typical school year start and end dates
    school_start_month, school_end_month = 9, 6  # September and June
    holiday_dates = {
        'Christmas': {'start': 12, 'end': 1, 'start_day': 24, 'end_day': 7},  # Christmas break
        'Easter': {'start': 3, 'end': 4, 'start_day': 15, 'end_day': 30},  # Easter break
        'Midterm_1': {'start': 10, 'end': 10, 'start_day': 28, 'end_day': 31},  # October break
        'Midterm_2': {'start': 2, 'end': 2, 'start_day': 14, 'end_day': 18},  # February break
    }

    # Initialize an empty DataFrame
    calendar_df = pd.DataFrame(columns=['Year', 'Date', 'Is_School_Day'])

    # Generate data for each year in the range
    for year in range(start_year, end_year + 1):
        # Generate dates for the whole year
        start_date = datetime(year, 1, 1)
        end_date = datetime(year, 12, 31)
        total_days = (end_date - start_date).days + 1
        dates = [start_date + timedelta(days=x) for x in range(total_days)]

        # Determine if each date is a school day
        for date in dates:
            is_school_day = True
            if date.month < school_start_month and date.month > school_end_month:
                is_school_day = False  # Summer break
            for holiday_name, holiday_info in holiday_dates.items():
                if date.month == holiday_info['start'] and date.day >= holiday_info['start_day']:
                    is_school_day = False
                if date.month == holiday_info['end'] and date.day <= holiday_info['end_day']:
                    is_school_day = False

            calendar_df = calendar_df.append({'Year': year, 'Date': date, 'Is_School_Day': is_school_day}, ignore_index=True)

    return calendar_df

# Assuming your existing data_df has years ranging from 2019 to 2023
school_calendar_df = create_school_calendar(2019, 2023)

# Display the first few rows of the created school calendar dataset
school_calendar_df.head()
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.graph_objs as go
import pandas as pd

# Load your data here, replace 'your_traffic_data.csv' with your file path
traffic = pd.read_csv('traffic_data.csv')

app = dash.Dash(__name__)

# Layout with multiple interactive components
app.layout = html.Div([
    html.H1('Traffic Data Visualization'),
    html.Div([
        html.Div([
            html.Label('Select Year:'),
            dcc.Dropdown(
                id='year-dropdown',
                options=[{'label': year, 'value': year} for year in traffic['Year_x'].unique()],
                value=traffic['Year_x'].min()
            ),
        ], style={'width': '48%', 'display': 'inline-block'}),
        html.Div([
            html.Label('Select Week Range:'),
            dcc.RangeSlider(
                id='week-range-slider',
                min=traffic['Week_Number'].min(),
                max=traffic['Week_Number'].max(),
                value=[traffic['Week_Number'].min(), traffic['Week_Number'].max()],
                marks={str(week): str(week) for week in traffic['Week_Number'].unique()},
                step=None
            ),
        ], style={'width': '48%', 'display': 'inline-block'}),
    ]),
    
    dcc.Graph(id='traffic-graph'),
    
    html.Div(id='hover-data', style={'padding': '20px'}),
])

# Callback for updating the graph
@app.callback(
    Output('traffic-graph', 'figure'),
    [Input('year-dropdown', 'value'),
     Input('week-range-slider', 'value')]
)
def update_graph(selected_year, week_range):
    filtered_df = traffic[(traffic['Year_x'] == selected_year) & 
                          (traffic['Week_Number'] >= week_range[0]) & 
                          (traffic['Week_Number'] <= week_range[1])]
    
    fig = go.Figure(data=[
        go.Bar(
            x=filtered_df['Week_Number'],
            y=filtered_df['VALUE'],  # Changed 'Volume_of_Cars' to 'VALUE'
            text=filtered_df['VALUE'],  # Changed 'Volume_of_Cars' to 'VALUE'
            hoverinfo='text+x+y'
        )
    ])
    
    fig.update_layout(
        title=f'Weekly Traffic Volume for {selected_year}',
        xaxis_title='Week Number',
        yaxis_title='Volume of Cars'  # You might want to change this title if 'VALUE' represents something else
    )
    
    return fig

# Callback for displaying hover data
@app.callback(
    Output('hover-data', 'children'),
    [Input('traffic-graph', 'hoverData')]
)
def display_hover_data(hoverData):
    if hoverData is not None:
        week = hoverData['points'][0]['x']
        volume = hoverData['points'][0]['y']
        return f'Week {week}: {volume} cars passed'  # Change 'cars passed' if 'VALUE' is not about cars
    return 'Hover over a bar to see details'

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html, Input, Output, dash_table
import plotly.graph_objs as go
import pandas as pd

# Load your data here
traffic1 = pd.read_csv('traffic_data.csv')

app = dash.Dash(__name__)

# Layout with multiple interactive components
app.layout = html.Div([
    html.H1('Traffic Data Visualization'),
    
    # Dropdown for selecting graph type
    html.Div([
        html.Label('Select Graph Type:'),
        dcc.Dropdown(
            id='graph-type-dropdown',
            options=[
                {'label': 'Bar Chart', 'value': 'bar'},
                {'label': 'Line Chart', 'value': 'line'}
            ],
            value='bar'
        ),
    ], style={'width': '30%', 'display': 'inline-block'}),

    # Dropdown for year
    html.Div([
        html.Label('Select Year:'),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': year, 'value': year} for year in traffic1['Year_x'].unique()],
            value=traffic1['Year_x'].min()
        ),
    ], style={'width': '30%', 'display': 'inline-block'}),

    # Week range slider
    html.Div([
        html.Label('Select Week Range:'),
        dcc.RangeSlider(
            id='week-range-slider',
            min=traffic1['Week_Number'].min(),
            max=traffic1['Week_Number'].max(),
            value=[traffic1['Week_Number'].min(), traffic1['Week_Number'].max()],
            marks={str(week): str(week) for week in traffic1['Week_Number'].unique()},
            step=None
        ),
    ], style={'width': '30%', 'display': 'inline-block'}),

    # Graph
    dcc.Graph(id='traffic-graph'),

    # Summary statistics
    html.Div(id='summary-stats', style={'padding': '20px'}),
    
    # Data table
    dash_table.DataTable(
        id='data-table',
        columns=[{"name": i, "id": i} for i in traffic.columns],
        data=traffic1.to_dict('records'),
    ),
    
    # Hover data
    html.Div(id='hover-data', style={'padding': '20px'}),
])

# Callback for updating the graph
@app.callback(
    Output('traffic-graph', 'figure'),
    [Input('graph-type-dropdown', 'value'),
     Input('year-dropdown', 'value'),
     Input('week-range-slider', 'value')]
)
def update_graph(graph_type, selected_year, week_range):
    filtered_df = traffic[(traffic1['Year_x'] == selected_year) & 
                          (traffic1['Week_Number'] >= week_range[0]) & 
                          (traffic1['Week_Number'] <= week_range[1])]
    
    if graph_type == 'bar':
        fig = go.Figure(data=[
            go.Bar(
                x=filtered_df['Week_Number'],
                y=filtered_df['VALUE'],
                text=filtered_df['VALUE'],
                hoverinfo='text+x+y'
            )
        ])
    elif graph_type == 'line':
        fig = go.Figure(data=[
            go.Scatter(
                x=filtered_df['Week_Number'],
                y=filtered_df['VALUE'],
                mode='lines+markers',
                hoverinfo='text+x+y'
            )
        ])
    
    fig.update_layout(
        title=f'Weekly Traffic Volume for {selected_year}',
        xaxis_title='Week Number',
        yaxis_title='Volume of Cars'
    )
    
    return fig

# Callback for updating summary statistics
@app.callback(
    Output('summary-stats', 'children'),
    [Input('year-dropdown', 'value'),
     Input('week-range-slider', 'value')]
)
def update_summary_stats(selected_year, week_range):
    filtered_df = traffic[(traffic1['Year_x'] == selected_year) & 
                          (traffic1['Week_Number'] >= week_range[0]) & 
                          (traffic1['Week_Number'] <= week_range[1])]
    average = filtered_df['VALUE'].mean()
    total = filtered_df['VALUE'].sum()
    return f'Average Volume: {average:.2f}, Total Volume: {total}'

# Callback for updating data table
@app.callback(
    Output('data-table', 'data'),
    [Input('year-dropdown', 'value'),
     Input('week-range-slider', 'value')]
)
def update_table(selected_year, week_range):
    filtered_df = traffic[(traffic1['Year_x'] == selected_year) & 
                          (traffic1['Week_Number'] >= week_range[0]) & 
                          (traffic1['Week_Number'] <= week_range[1])]
    return filtered_df.to_dict('records')

# Callback for displaying hover data
@app.callback(
    Output('hover-data', 'children'),
    [Input('traffic-graph', 'hoverData')]
)
def display_hover_data(hoverData):
    if hoverData is not None:
        week = hoverData['points'][0]['x']
        volume = hoverData['points'][0]['y']
        return f'Week {week}: {volume} cars passed'
    return 'Hover over a bar to see details'

if __name__ == '__main__':
    app.run_server(debug=True)
```
